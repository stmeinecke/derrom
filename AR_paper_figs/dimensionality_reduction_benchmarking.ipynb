```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [9, 6]

%matplotlib notebook
```

```python
svd_std = np.loadtxt('/home/agluedge/Preprints/ELPH/figs/red_dim_apprx/svd_std_2').T
svd_max = np.loadtxt('/home/agluedge/Preprints/ELPH/figs/red_dim_apprx/svd_max_2').T

FFT_std = np.loadtxt('/home/agluedge/Preprints/ELPH/figs/red_dim_apprx/FFT_std_2').T
FFT_max = np.loadtxt('/home/agluedge/Preprints/ELPH/figs/red_dim_apprx/FFT_max_2').T

hermite_std = np.loadtxt('/home/agluedge/Preprints/ELPH/figs/red_dim_apprx/hermite_std_2').T
hermite_max = np.loadtxt('/home/agluedge/Preprints/ELPH/figs/red_dim_apprx/hermite_max_2').T
```

```python
####################################################
####################################################
####################################################

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif"
})

from matplotlib import gridspec
import matplotlib.ticker as ticker

fs = 9
fig = plt.figure(figsize=(3.40457,3.5))
plt.rcParams.update({'font.size': fs})
gs = gridspec.GridSpec(nrows=2, ncols=1, width_ratios=[1], height_ratios=[1,1])


ms = 5

xmin = 50
xmax = 250



####################################################
####################################################
####################################################


ax00 = plt.subplot(gs[0,0])

plt.grid(0.5)
plt.plot(svd_std[0], svd_std[1]/svd_std[0], '--', marker='o', label='SVD', ms=5)
plt.plot(FFT_std[0], FFT_std[1]/FFT_std[0], '--', marker='o', label='DFT', ms=5)
plt.plot(hermite_std[0], hermite_std[1]/hermite_std[0], '--', marker='o', label='Hermite', ms=5)

# plt.xlim(svd_std[0,0],svd_std[0,-1])
plt.xlim(xmin, xmax)
plt.ylim(0,1.1)

# plt.legend(loc='right')

plt.title('rms error', fontsize=fs)
plt.xlabel(r'discretization points $d_s$')
plt.ylabel(r'compression ratio $d_r / d_s$')

ax00.xaxis.set_ticks_position('both')
ax00.yaxis.set_ticks_position('both')
ax00.tick_params(which='both', direction='in', color='k')

ax00.xaxis.set_major_locator(ticker.MultipleLocator(50))
ax00.xaxis.set_minor_locator(ticker.MultipleLocator(10))
# ax00.xaxis.set_major_formatter(ticker.NullFormatter())
# ax00.xaxis.set_minor_formatter(ticker.NullFormatter())

ax00.yaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax00.yaxis.set_minor_locator(ticker.MultipleLocator(0.04))
#ax00.yaxis.set_major_formatter(ticker.NullFormatter())
#ax00.yaxis.set_minor_formatter(ticker.NullFormatter())

plt.text(0.03, 0.97,r'(a)', horizontalalignment='left', verticalalignment='top', transform=ax00.transAxes)

####################################################
####################################################
####################################################


ax00 = plt.subplot(gs[1,0])

plt.grid(0.5)
plt.plot(svd_max[0], svd_max[1]/svd_max[0], '--', marker='o', label='SVD', ms=5)
plt.plot(FFT_max[0], FFT_max[1]/FFT_max[0], '--', marker='o', label='FFT', ms=5)
plt.plot(hermite_max[0], hermite_max[1]/hermite_max[0], '--', marker='o', label='Hermite', ms=5)

# plt.xlim(svd_std[0,0],svd_std[0,-1])
plt.xlim(xmin, xmax)
plt.ylim(0,1.1)

plt.legend()

plt.title('maximum error', fontsize=fs)
plt.xlabel(r'discretization points $d_s$')
plt.ylabel(r'compression ratio $d_r / d_s$')

ax00.xaxis.set_ticks_position('both')
ax00.yaxis.set_ticks_position('both')
ax00.tick_params(which='both', direction='in', color='k')

ax00.xaxis.set_major_locator(ticker.MultipleLocator(50))
ax00.xaxis.set_minor_locator(ticker.MultipleLocator(10))
# ax00.xaxis.set_major_formatter(ticker.NullFormatter())
# ax00.xaxis.set_minor_formatter(ticker.NullFormatter())

ax00.yaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax00.yaxis.set_minor_locator(ticker.MultipleLocator(0.04))
#ax00.yaxis.set_major_formatter(ticker.NullFormatter())
#ax00.yaxis.set_minor_formatter(ticker.NullFormatter())

plt.text(0.03, 0.97,r'(b)', horizontalalignment='left', verticalalignment='top', transform=ax00.transAxes)

####################################################
####################################################
####################################################


plt.tight_layout()

plt.subplots_adjust(left=0.12, bottom=0.10, right=0.96, top=0.95, wspace=0.08, hspace=0.55)

plt.savefig('/home/agluedge/Preprints/ELPH/figs/dimensionality_reduction.pdf')
plt.show()
```
