```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [9, 6]

# %matplotlib notebook


import sys
sys.path.append("../incl/")
```

```python
import os
path = '/home/agluedge/Preprints/ELPH/data/ELM_features_rdim/'

filenames = []
# r=root, d=directories, f = filenames
for r, d, f in os.walk(path):
    for file in f:
        if 'rdim_' in file:
            filenames.append(os.path.join(r, file))

#for f in filenames:
    #print(f)

from natsort import realsorted
filenames = realsorted(filenames)


# ##print filenames
# for f in filenames:
#     print(f)
# nYdata = len(filenames)
# print(nYdata)
    
  
scans = [] # that data goes into this list
for fn in filenames:
  scans.append(np.loadtxt(fn)) #load linescan an append to the list of linescans - each linescan is matrix

scans = np.asarray(scans).T #create numpy ndarray from list -- contains all the data -- format [number of files, sweep steps, data column]
print (scans.shape)
```

```python
rdimAr = range(5,5+scans.shape[1],1)

rmax = 30

plt.grid()
plt.errorbar(rdimAr[:rmax], scans[0,:rmax], yerr=scans[1,:rmax])
plt.show()

plt.grid()
plt.errorbar(rdimAr[:rmax], scans[2,:rmax], yerr=scans[3,:rmax])
plt.show()

plt.plot(rdimAr[:rmax], scans[4,:rmax], label='std')
plt.plot(rdimAr[:rmax], scans[5,:rmax], label='max')
plt.legend()
plt.show()
```
